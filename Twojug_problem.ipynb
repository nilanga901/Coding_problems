```python

# Two jugs J1 and J2 with capacities C1 and C2 (infinite water)

# Measure T units of water (T <= (C1 + C2))

# No markings on jugs i.e., we cannot measure a partially filled jug directly


# Actions available

# Fill up J1 completely

# Fill up J2 completely

# Empty J1 completely

# Empty J2 completely

# Transfer water from J1 to J2 

# Transfer water from J2 to J1

# Sample Example

# C1 = 5, C2 = 3, T = 2


# Start with (0, 0)

# State -> amount of water contained in the jugs

# Empty, Fill, Transfer for each Jug



# (0, 0) -> (5, 0) -> (2, 3) -> (2, 0)

# C1 = 5, C2 = 3, T = 4

#T=x1+x2

# (0, 0) -> (5, 0) -> (2, 3) -> (2, 0) -> (0,2) ->(5,2)->(4,3)->(4,0)



# Test Cases
# Case 1
# C1 = 5, C2 = 3, T = 4
# Output: true

# Case 2
# C1 = 2, C2 = 6, T = 5
# Output: false

# Case 3
# C1 = 1, C2 = 2, T = 3
# Output: true
```

```python

def transition(state, action,C1,C2):
    state=list(state)
    if(action==0):
        state[0]=C1
    elif(action==1):
        state[1]=C2
    elif(action==2):
        state[0]=0
    elif(action==3):
        state[1]=0
    elif(action==4):
        transfer_amount=min(state[0],C2-state[1])
        state[0]-=transfer_amount
        state[1]+=transfer_amount
    elif(action==5):
        transfer_amount=min(state[1],C1-state[0])
        state[1]-=transfer_amount
        state[0]+=transfer_amount
    return (state[0],state[1])


def find_optim_path(C1,C2,T):

    start=(0,0)
    #Fill up J1 completely -0
    # Fill up J2 completely 1
    # Empty J1 completely -2
    # Empty J2 completely -3
    # Transfer water from J1 to J2  -4
    # Transfer water from J2 to J1 -5
        
    
    found=False;
    nodelist=[start]
    visited=set()
    action_seq={start:[]}

    while(nodelist):
        print(nodelist)
        current=nodelist.pop(0)
        visited.add(current)

        for act in range(6):
            new_state=transition(current,act,C1,C2)
            if(new_state not in visited):
                action_seq[new_state]=action_seq[current]+[act]   #keeptrack_of_action_sequence
                nodelist.append(new_state)
                if(sum(new_state)==T):
                    print(new_state)
                    found=True
                    break
            
        if(found):
            break
        
    return (found,action_seq[new_state])
        
```
